```python
# Import the dependencies.
# - Pandas to read our CSV file and create the locations and measurements from the DataFrame. 
# - gmaps and the API key to create heatmaps and the locations map,
# - requests to make a request to the Google Places JSON file. 
# - allows us to get hotel locations from the latitude and longitude of the city.

import pandas as pd
import gmaps
import requests
# Import the Goggle Maps API key.
from config import g_key
```

```python
# 2. Import the WeatherPy_database.csv file from Part 1 as a new DataFrame.
# Read & store the CSV you saved created in part one into a DataFrame.
city_weather_data_df = pd.read_csv("weather_data/cities_weather.csv", index_col='City_ID')
city_weather_data_df
```

```python
# 3. Filter the DataFrame for minimum and maximum temperature preferences, 
# - and if the rain or snow accumulation is 0 inches or not using conditional statements. 
# - Do the following:
# -- Prompt the customer for the minimum temperature preference.
# -- Prompt the customer for the maximum temperature preference.
# -- Prompt the customer to answer if he or she would like it to be raining or not, using input("Do you want it to be raining? (yes/no) ").
# -- Prompt the customer to answer if he or she would like it to be snowing or not, using input("Do you want it to be snowing? (yes/no) ").
min_temp = float(input("What is the minimum temperature you would like for your trip? "))
max_temp = float(input("What is the maximum temperature you would like for your trip? "))
rain_yn = input("Do you want it to be raining? (yes/no) ")
snow_yn = input("Do you want it to be snowing? (yes/no) ")

# Filter the dataset to find the cities that fit the criteria.
if rain_yn == 'y':
    raining = city_weather_data_df['Rain last hour'] > 0
else:
    raining = city_weather_data_df['Rain last hour'] == 0
if snow_yn == 'y':
    snowing = city_weather_data_df['Snow last hour'] > 0
else:
    snowing = city_weather_data_df['Snow last hour'] == 0

preferred_cities_df = city_weather_data_df.loc[
    (city_weather_data_df["Max Temp"] <= max_temp) &
    (city_weather_data_df["Max Temp"] >= min_temp) &
    raining &
    snowing
]
preferred_cities_df
```

```python
# 4. Add the cities to a marker layer map with a pop-up marker for each city that includes:
# -- Hotel name
# -- City
# -- Country
# -- Current weather description 
# -- with the maximum temperature

# Find the nearest hotel for each city using a Nearby Search, use these parameters:
# - API key
# - Latitude and longitude
# - 5,000-meter radius
# - Type of place
params = {
    "radius": 5000,
    "type": "lodging",
    "key": g_key
}

# Iterate through preferred_cities_df DataFrame
# Use iterrows() function to perform the iteration, provide the index and the row in the for loop
for index, row in preferred_cities_df.iterrows():
    # Get the latitude and longitude.
    lat = row["Lat"]
    lng = row["Lng"]

    # Add the latitude and longitude to location key for the params dictionary.
    params["location"] = f"{lat},{lng}"

    # Use the search term: "lodging" and our latitude and longitude.
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    # Make request and get the JSON data from the search.
    hotels = requests.get(base_url, params=params).json()
    # Grab the first hotel from the results and store the name.
    try:
        print(hotels["results"][0]["name"])

        preferred_cities_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    except (IndexError):
        print("Hotel not found, set as UNKNOWN... skipping.")
        preferred_cities_df.loc[index, "Hotel Name"] = 'UNKNOWN'

        # If you encounter more errors, add them to the except block—like so, (IndexError, NewError)—and continue running your code until there are no errors.

preferred_cities_df
```

```python
# Tell gmaps to use your API key, needed only once.
gmaps.configure(api_key=g_key)
# Add a marker for each hotel.
locations = preferred_cities_df[["Lat", "Lng"]]
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
marker_layer = gmaps.marker_layer(locations)
fig.add_layer(marker_layer)
# Call the figure to plot the data.
fig
```

```python
# Add popup marker for each city
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather Description</dt><dd>{Weather}</dd>
<dt>Max Temp</dt><dd>{Max Temp} °F</dd>
</dl>
"""

# Store the DataFrame Row.
city_info = [info_box_template.format(**row) for index, row in preferred_cities_df.iterrows()]
display(city_info)

# Add a pop-up marker for each city.
locations = preferred_cities_df[["Lat", "Lng"]]
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
marker_layer = gmaps.marker_layer(locations, info_box_content=city_info)
fig.add_layer(marker_layer)

# Call the figure to plot the data.
fig
```

```python
# 5. Save and upload the new DataFrame as WeatherPy_vacation.csv.
output_data_file = "weather_data/WeatherPy_Vacation.csv"
# Export the city_weather_data into a CSV.
preferred_cities_df.to_csv(output_data_file, index_label="City_ID")
```
